### **2b. count_total_students(filepath="records.csv")**
### **Key considerations(Thought Process):**
#### We are considering to get through the whole list of students in each tutorial group as stated in the csv file, so the key assumption is that each tutorial group all have 50 students.

#### This function is to find the total number of students that is in the csv file, so that we can decide how many loops we should consider to finish running the entire csv file to form groups



1. Open "records.csv" with **proper encoding and newline handling**
2. Create CSV reader iterator
3. Skip header row (if present)
4. For each remaining row: generate 1
5. Sum all generated 1s → total student count
6. Return count 

### **Return Value:**
##### Returns an integer representing the total number of student records in the CSV file, excluding the header row.



````python
import csv
import itertools
import os
import random
import sys


def read_students(start_index=0, count=50, filepath="records.csv"):
    """Read up to count students starting from start_index (after header).
    Returns a list of student dicts with keys: tutorial group, id, school, name, gender, gpa
    """
    students = []

    with open(filepath, newline='', encoding='utf-8') as f:
        reader = csv.reader(f)
        headers = next(reader, None)  # skip header if present

        # skip start_index rows
        for _ in range(start_index):
            try:
                next(reader)
            except StopIteration:
                return students

        for _ in range(count):
            try:
                row = next(reader)
                
            except StopIteration:
                break

            try:
                gpa = float(row[5])
            except Exception:
                # skip rows with non-numeric GPA
                continue

            student = {
                'tutorial group': row[0].strip(),
                'id': row[1].strip(),
                'school': row[2].strip(),
                'name': row[3].strip(),
                'gender': row[4].strip(),
                'gpa': gpa,
            }
            students.append(student)

    return students
    
read_students(start_index=0, count=50, filepath="records.csv")
````

````python

def count_total_students(filepath="records.csv"):
    """Count total number of student records (excluding header)."""        
    with open(filepath, newline='', encoding='utf-8') as f:
        reader = csv.reader(f)
        headers = next(reader, None)  # skip header
        return sum(1 for row in reader)
        
count_total_students(filepath="records.csv")
````

````python
def form_groups(students, num_groups=10):
    """Form groups from a list of student dicts.
    Returns list of groups (each group is a list of students).
    """
    if not students:
        return []

    students_sorted = sorted(students, key=lambda x: x['gpa'], reverse=True)

    # split into up to 5 bands as evenly as possible
    n = len(students_sorted)
    band_size = n // 5
    bands = [students_sorted[i:i + band_size] for i in range(0, n, band_size)]

    # split each band by gender
    band_males = []
    band_females = []
    for band in bands:
        males = [s for s in band if s['gender'].strip().lower() in ('m', 'male')]
        females = [s for s in band if s['gender'].strip().lower() in ('f', 'female')]
        random.shuffle(males)
        random.shuffle(females)
        band_males.append(males)
        band_females.append(females)

    total_males = sum(len(m) for m in band_males)
    total_females = sum(len(f) for f in band_females)
    
    if total_males > total_females:
        target_males_per_group, target_females_per_group = 3, 2
    else:
        target_males_per_group, target_females_per_group = 2, 3

    groups = [[] for _ in range(num_groups)]
    male_count = [0] * num_groups
    female_count = [0] * num_groups

    # distribute one student from each band into each group
    for band_idx in range(len(bands)):
        for group_idx in range(num_groups):
            # try to pick a student to respect the gender targets
            picked = False
            if male_count[group_idx] < target_males_per_group and band_males[band_idx]:
                groups[group_idx].append(band_males[band_idx].pop())
                male_count[group_idx] += 1
                picked = True
            elif female_count[group_idx] < target_females_per_group and band_females[band_idx]:
                groups[group_idx].append(band_females[band_idx].pop())
                female_count[group_idx] += 1
                picked = True
            elif band_males[band_idx]:
                groups[group_idx].append(band_males[band_idx].pop())
                male_count[group_idx] += 1
                picked = True
            elif band_females[band_idx]:
                groups[group_idx].append(band_females[band_idx].pop())
                female_count[group_idx] += 1
                picked = True

    return groups 
````

### Final Output Structure
The end result shows a clear, organized view where you can easily see:

- All tutorial groups listed separately

- Under each tutorial group, which teams contain its students

- Detailed information about each student in those teams

- Team-level statistics for quick reference

This approach makes it simple to understand how students from each tutorial group are distributed across the various teams while maintaining all relevant student information and team statistics.


````python
def print_groups(groups, students_sorted, bands, batch_number=1):
    """Print groups organized by Tutorial Group with team allocations."""
    print(f"\n=== Team Assignments by Tutorial Group ===")
    
    # Create band lookup for this batch
    student_to_band = {}
    for band_idx, band in enumerate(bands):
        for student in band:
            student_key = f"{student['name']}_{student['id']}"
            student_to_band[student_key] = band_idx + 1
    
    # Create team assignment lookup from the groups
    student_to_team = {}
    for team_idx, team in enumerate(groups, 1):
        for student in team:
            student_key = f"{student['name']}_{student['id']}"
            student_to_team[student_key] = team_idx
    
    # Organize all students by tutorial group
    tutorial_groups = {}
    for student in students_sorted:
        tutorial_group = student['tutorial group']
        if tutorial_group not in tutorial_groups:
            tutorial_groups[tutorial_group] = []
        tutorial_groups[tutorial_group].append(student)
    
    # Print each tutorial group with team assignments
    for tutorial_group in sorted(tutorial_groups.keys()):
        print(f"\nTutorial Group: {tutorial_group}")
        
        students_in_tutorial = tutorial_groups[tutorial_group]
        
        # Organize students by their assigned teams
        team_assignments = {}
        for student in students_in_tutorial:
            student_key = f"{student['name']}_{student['id']}"
            team_num = student_to_team.get(student_key)
            if team_num is not None:
                if team_num not in team_assignments:
                    team_assignments[team_num] = []
                team_assignments[team_num].append(student)
        
        # Print students for each team
        for team_num in sorted(team_assignments.keys()):
            team_students = team_assignments[team_num]
            print(f"  Team {team_num}:")
            
            total_gpa = 0.0
            male_count = 0
            female_count = 0
            
            for student_num, student in enumerate(team_students, 1):
                student_key = f"{student['name']}_{student['id']}"
                band_number = student_to_band.get(student_key, "Unknown")
                
                gender_lower = student['gender'].lower().strip()
                if gender_lower in ('m', 'male'):
                    gender_display = 'M'
                    male_count += 1
                elif gender_lower in ('f', 'female'):
                    gender_display = 'F'
                    female_count += 1
                else:
                    gender_display = student['gender']
                
                print(f"    {student_num}. {student['name']} "
                      f"(ID: {student['id']}, "
                      f"School: {student['school']}, "
                      f"Gender: {gender_display}, "
                      f"CGPA: {student['gpa']:.2f}, "
                      f"Band: {band_number})")
                
                total_gpa += student['gpa']
            
            # Print team summary
            if team_students:
                avg_gpa = total_gpa / len(team_students)
                print(f"    Summary: {len(team_students)} students "
                      f"(M: {male_count}, F: {female_count}), "
                      f"Avg GPA: {avg_gpa:.2f}")
        
        # Print tutorial group summary
        total_in_tutorial = len(students_in_tutorial)
        assigned_count = sum(len(students) for students in team_assignments.values())
        if assigned_count < total_in_tutorial:
            print(f"  Note: {total_in_tutorial - assigned_count} student(s) not assigned to any team")
````

````python
def save_groups_to_csv(all_groups_data, output_file="student_groups.csv"):
    """Save all groups data to a CSV file.
    
    Args:
        all_groups_data: List of tuples (batch_number, groups, students_sorted, bands)
        output_file: Name of the output CSV file
    """
    try:
        with open(output_file, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            
            # Write header with required column names
            writer.writerow([
                'Tutorial Group', 'Student ID', 'School', 'Name', 'Gender', 'CGPA', 'Team Assigned'
            ])
            
            # Precompute band assignments for each batch and write data
            for batch_number, groups, students_sorted, bands in all_groups_data:
                # Create band lookup for this batch
                student_to_band = {}
                for band_idx, band in enumerate(bands):
                    for student in band:
                        student_key = f"{student['name']}_{student['id']}"
                        student_to_band[student_key] = band_idx + 1
                
                # Write groups for this batch
                for group_idx, group in enumerate(groups, 1):
                    for student in group:
                        writer.writerow([
                            student['tutorial group'],
                            student['id'],
                            student['school'],
                            student['name'],
                            student['gender'],
                            f"{student['gpa']:.2f}",
                            f"Team {group_idx}"
                        ])
        
        print(f"All groups have been saved to '{output_file}'")
        return True
        
    except Exception as e:
        print(f"Error saving to CSV file: {e}")
        return False
````

### 5c. Execution of all the functions

#### Batch Processing:

1. Reads 50 students at a time instead of loading all data at once --> Prevents memory issues with large student datasets

#### Academic Banding:

1. Sorts students by GPA (highest first)
2. Divides them into 5 performance "bands" (top 20%, next 20%, etc.)
3. Ensures groups get students from different performance levels

#### Group Formation:

1. Creates 10 balanced groups per batch using form_groups() functio
2. Each group gets a mix of high, medium, and low-performing students


#### Output:
1. Prints groups for each batch to console
2. Saves ALL group data to a CSV file named 'students_group.csv' after processing all batches

````python
if __name__ == "__main__":
    BATCH_SIZE = 50
    NUM_GROUPS = 10
    all_groups_data = []  # ← ADD THIS LINE
    
    try:
        # Count total students first
        total_students = count_total_students()
        if total_students == 0:
            print("No student records found in the file.")
            sys.exit(1)

        
        batch_number = 1
        start_index = 0
        
        # Process file in batches until all students are processed
        while True:
            print(f"Reading batch {batch_number} (students {start_index + 1} to {start_index + BATCH_SIZE})...")
            students = read_students(start_index, BATCH_SIZE)
            
            if not students:
                print("No more students to process.")
                break
                
            print(f"Processing {len(students)} students {batch_number} times...")
            
            # Sort students for band creation
            students_sorted = sorted(students, key=lambda x: x['gpa'], reverse=True)
            n = len(students_sorted)
            band_size = max(1, n // 5)
            bands = [students_sorted[i:i + band_size] for i in range(0, n, band_size)]

            # Form groups for this batch
            groups = form_groups(students, num_groups=NUM_GROUPS)
            
            # Print groups for this batch
            print_groups(groups, students_sorted, bands, batch_number)

            all_groups_data.append((batch_number, groups, students_sorted, bands))
            
            # Update for next batch
            start_index += BATCH_SIZE
            batch_number += 1
            
            # Add separator between batches
            if start_index < total_students:
                print("=" * 60)
                print()
        
        # MOVE CSV SAVING HERE - AFTER ALL BATCHES ARE PROCESSED
        if all_groups_data:
            save_groups_to_csv(all_groups_data, "student_groups.csv")
            print(f"All groups have been saved to 'student_groups.csv'")
        else:
            print("No group data to save.")
        
        print(f"Processing complete! Processed {batch_number - 1} batch(es) total.")
        
    except FileNotFoundError as e:
        print(e)
        sys.exit(1)
        
    except Exception as e:
        print(f"An error occurred: {e}")
        sys.exit(1)
````

````python
import matplotlib.pyplot as plt
import csv
import random

# Read data from records.csv file
tutorial_groups = []
gpas = []
teams = []

with open('FDAB_Team5_XuHanwen.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        tutorial_groups.append(row['Tutorial Group'])
        gpas.append(float(row['CGPA']))  # Using CGPA column for GPA
        teams.append(row['Team Assigned'])  # Using 'Team Assigned' column

# Organize data by tutorial group and team
group_team_data = {}
for i, group in enumerate(tutorial_groups):
    if group not in group_team_data:
        group_team_data[group] = {}
    team = teams[i]
    if team not in group_team_data[group]:
        group_team_data[group][team] = []
    group_team_data[group][team].append(gpas[i])

# Calculate statistics
sorted_groups = sorted(group_team_data.keys(), key=lambda x: int(x.split('-')[1]))
group_means = []
team_means_all = []  # Store all team averages
all_student_gpas = []

# Prepare data for plotting
x_positions_teams = []  # For team averages (blue dots)
y_gpas_teams = []       # For team averages (blue dots)
x_positions_groups = [] # For group averages (red dots)
y_gpas_groups = []      # For group averages (red dots)
group_labels = []

for i, group in enumerate(sorted_groups):
    group_gpas = []
    team_averages = []
    
    # Calculate team averages within this tutorial group
    for team, team_gpas in group_team_data[group].items():
        team_avg = sum(team_gpas) / len(team_gpas)
        team_averages.append(team_avg)
        team_means_all.append(team_avg)
        
        # Add team average as blue dot
        x_positions_teams.append(i)
        y_gpas_teams.append(team_avg)
        
        # Collect all individual GPAs for overall statistics
        group_gpas.extend(team_gpas)
        all_student_gpas.extend(team_gpas)
    
    # Calculate group average (average of team averages)
    if team_averages:
        group_avg = sum(team_averages) / len(team_averages)
    else:
        group_avg = 0
    
    group_means.append(group_avg)
    
    # Add group average as red dot
    x_positions_groups.append(i)
    y_gpas_groups.append(group_avg)
    group_labels.append(group)

# Create the visualization
plt.figure(figsize=(20, 10))

# Plot team average GPAs as blue dots
plt.scatter(x_positions_teams, y_gpas_teams, alpha=0.7, color='blue', s=50, label='Team Average GPA')

# Plot tutorial group average GPAs as red dots and connect with line
plt.plot(x_positions_groups, y_gpas_groups, 'ro-', 
         linewidth=2, markersize=8, label='Tutorial Group Average GPA')

# Customize the plot
plt.xlabel('Tutorial Groups', fontsize=15)
plt.ylabel('GPA', fontsize=15)
plt.title('GPA Distribution Across Tutorial Groups\n(Team Averages and Group Averages)', fontsize=14)

# Set x-axis labels to show every 10th tutorial group for readability
x_ticks = list(range(0, len(sorted_groups), 10))
x_labels = [sorted_groups[i] for i in x_ticks]
plt.xticks(x_ticks, x_labels, rotation=45)

# Set y-axis range from 3.6 to 4.6
plt.ylim(3.6, 4.6)

# Add grid and legend
plt.grid(True, alpha=0.3)
plt.legend()

# Add horizontal line for overall average
overall_mean = sum(all_student_gpas) / len(all_student_gpas)
plt.axhline(y=overall_mean, color='green', linestyle='--', 
            linewidth=2, alpha=1.0, label=f'Overall Student Average: {overall_mean:.2f}')

# Add horizontal line for overall team average
overall_team_mean = sum(team_means_all) / len(team_means_all) if team_means_all else 0
plt.axhline(y=overall_team_mean, color='orange', linestyle='--', 
            linewidth=2, alpha=1.0, label=f'Overall Team Average: {overall_team_mean:.2f}')

plt.legend()
plt.tight_layout()
plt.show()

# Print statistical summary
print("=" * 60)
print("STATISTICAL SUMMARY")
print("=" * 60)
print(f"Total Students: {len(all_student_gpas)}")
print(f"Total Tutorial Groups: {len(sorted_groups)}")
print(f"Total Teams: {len(team_means_all)}")
print(f"Overall Student Average GPA: {overall_mean:.3f}")
print(f"Overall Team Average GPA: {overall_team_mean:.3f}")
print(f"Minimum Student GPA: {min(all_student_gpas):.3f}")
print(f"Maximum Student GPA: {max(all_student_gpas):.3f}")

# Print group statistics
print(f"\nTutorial Group GPA Statistics:")
print(f"Average group mean GPA: {sum(group_means)/len(group_means):.3f}")
print(f"Lowest group average: {min(group_means):.3f}")
print(f"Highest group average: {max(group_means):.3f}")

# Print team statistics
print(f"\nTeam GPA Statistics:")
print(f"Average team mean GPA: {sum(team_means_all)/len(team_means_all):.3f}")
print(f"Lowest team average: {min(team_means_all):.3f}")
print(f"Highest team average: {max(team_means_all):.3f}")

# Show some sample groups with their averages
print(f"\nSample Group Averages:")
for i in range(min(5, len(sorted_groups))):
    print(f"  {sorted_groups[i]}: {group_means[i]:.3f}")
if len(sorted_groups) > 5:
    print("  ...")
    for i in range(len(sorted_groups)-3, len(sorted_groups)):
        print(f"  {sorted_groups[i]}: {group_means[i]:.3f}")

# Show some sample teams with their averages
print(f"\nSample Team Averages:")
sample_teams_shown = 0
for i, group in enumerate(sorted_groups):
    if sample_teams_shown >= 8:  # Show 8 sample teams total
        break
    for team, team_gpas in group_team_data[group].items():
        if sample_teams_shown >= 8:
            break
        team_avg = sum(team_gpas) / len(team_gpas)
        print(f"  {group} - Team {team}: {team_avg:.3f}")
        sample_teams_shown += 1
````

````python
import matplotlib.pyplot as plt
import csv
from collections import defaultdict
import random

# Read data from records.csv file
tutorial_team_school_counts = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))

with open('FDAB_Team5_XuHanwen.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        tutorial_group = row['Tutorial Group']
        team = row['Team Assigned']
        school = row['School']
        tutorial_team_school_counts[tutorial_group][team][school] += 1

# Get all unique schools and randomly select 10 tutorial groups
all_schools = set()
all_tutorial_groups = list(tutorial_team_school_counts.keys())

# Randomly select 10 tutorial groups
random.seed(42)  # For reproducible results
selected_tutorial_groups = random.sample(all_tutorial_groups, min(10, len(all_tutorial_groups)))

# Collect all schools from selected tutorial groups
for tutorial_group in selected_tutorial_groups:
    for team in tutorial_team_school_counts[tutorial_group]:
        all_schools.update(tutorial_team_school_counts[tutorial_group][team].keys())

all_schools = sorted(list(all_schools))

# Prepare data for stacked bar chart
team_labels = []
school_data = {school: [] for school in all_schools}

# Create labels and data for each team in selected tutorial groups
for tutorial_group in selected_tutorial_groups:
    teams_in_group = sorted(tutorial_team_school_counts[tutorial_group].keys())
    for team in teams_in_group:
        team_label = f"{tutorial_group}\nTeam {team}"
        team_labels.append(team_label)
        
        total_in_team = sum(tutorial_team_school_counts[tutorial_group][team].values())
        
        for school in all_schools:
            count = tutorial_team_school_counts[tutorial_group][team].get(school, 0)
            school_data[school].append(count)

# Limit to maximum 5 teams per tutorial group (if needed)
max_teams_per_group = 5
filtered_team_labels = []
filtered_school_data = {school: [] for school in all_schools}

current_tutorial_group = None
teams_in_current_group = 0

for i, team_label in enumerate(team_labels):
    tutorial_group = team_label.split('\n')[0]
    
    if tutorial_group != current_tutorial_group:
        current_tutorial_group = tutorial_group
        teams_in_current_group = 0
    
    if teams_in_current_group < max_teams_per_group:
        filtered_team_labels.append(team_label)
        for school in all_schools:
            filtered_school_data[school].append(school_data[school][i])
        teams_in_current_group += 1

team_labels = filtered_team_labels
school_data = filtered_school_data

# Create the stacked bar chart
plt.figure(figsize=(20, 10))

# Create stacked bars
bottom = [0] * len(team_labels)
colors = plt.cm.Set3(range(len(all_schools)))

for i, school in enumerate(all_schools):
    plt.bar(team_labels, school_data[school], bottom=bottom, 
            label=school, color=colors[i], alpha=0.8, edgecolor='grey')
    bottom = [bottom[j] + school_data[school][j] for j in range(len(team_labels))]

# Customize the plot
plt.xlabel('Teams in Randomly Selected Tutorial Groups', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.title('Distribution of Schools Within Teams (10 Random Tutorial Groups, Max 5 Teams Each)', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Add grid
plt.grid(True, alpha=0.3, axis='y')

# Limit y-axis to 5
plt.ylim(0, 5)

plt.tight_layout()
plt.show()

# Print statistical summary
print("=" * 60)
print("STATISTICAL SUMMARY")
print("=" * 60)
print(f"Selected Tutorial Groups: {len(selected_tutorial_groups)}")
print(f"Selected Groups: {', '.join(sorted(selected_tutorial_groups))}")
print(f"Total Schools: {len(all_schools)}")
print(f"Schools: {', '.join(all_schools)}")

print(f"\nSchool Distribution Summary:")
for school in all_schools:
    total_students = sum(school_data[school])
    teams_with_students = sum(1 for count in school_data[school] if count > 0)
    print(f"  {school}: {total_students} students across {teams_with_students} teams")

# Show detailed distribution for selected tutorial groups
print(f"\nDetailed Team Distribution:")
for tutorial_group in selected_tutorial_groups:
    print(f"\n{tutorial_group}:")
    teams_in_group = sorted(tutorial_team_school_counts[tutorial_group].keys())
    teams_shown = 0
    for team in teams_in_group:
        if teams_shown >= 5:  # Limit to 5 teams per group
            remaining_teams = len(teams_in_group) - 5
            if remaining_teams > 0:
                print(f"  ... and {remaining_teams} more teams")
            break
            
        print(f"  Team {team}:")
        team_total = sum(tutorial_team_school_counts[tutorial_group][team].values())
        for school in all_schools:
            count = tutorial_team_school_counts[tutorial_group][team].get(school, 0)
            if count > 0:
                print(f"    {school}: {count} student(s)")
        teams_shown += 1
````
